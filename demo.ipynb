```python
# Import necessary libraries
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Move model to GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# Function to generate article
def generate_article(title, keyword):
    # Prepare the input prompt
    prompt = f"Write an popular article titled '{title}' dealing with the keywords '{keyword}'. The article will start with the given title. Then include helpful tips, a step-by-step guide, and make sure it's engaging and SEO-friendly."

    # Tokenize input and move to the same device as the model
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    # Create an attention mask
    attention_mask = torch.ones(inputs['input_ids'].shape, dtype=torch.long).to(device)


    # Generate article with sampling
    outputs = model.generate(
        inputs['input_ids'],
        max_length=800,  # Increase length for more content
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id,  # Properly pad
        attention_mask=attention_mask,
        temperature=0.9,  # Adjust randomness
        top_p=0.95,  # Use nucleus sampling
        do_sample=True  # Enable sampling for varied results
    )

    # Decode the generated text
    article = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return article

# Example Usage
title = "How to Start a Small Business at Home?"
keyword = "Home-based small business startup"

generated_article = generate_article(title, keyword)
print(generated_article)
```

```python
# Function to save the article to a file
def save_article_to_file(article, title):
    filename = f"{title.replace(' ', '_')}.txt"  # Create a filename by replacing spaces with underscores
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(article)
    return filename

# Save the article to a file
saved_file = save_article_to_file(generated_article, title)
print(f"Article saved to {saved_file}")
```
